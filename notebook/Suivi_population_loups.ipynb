```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_url = "https://raw.githubusercontent.com/dthonon/pna_loup_suivi/main/data/"
data_url = "/home/daniel/Code/pna_loup_suivi/data/"
# Lecture des estimations de population de loups
population = pd.read_csv(data_url + "nbloupsestime.csv", sep=";", decimal=",")
population
```

```python
population["pct_tués"] = population["nb_loups_tués"] / population["Esti_OFB"] * 100
population["pct_diff"] = population["Esti_OFB"].pct_change(fill_method=None) * 100
population[["Année", "nb_loups_tués", "pct_tués", "Esti_OFB", "pct_diff"]]
```

```python
# Apply the default theme
sns.set_theme()
sns.set(rc={'figure.figsize':(9.05,2.8)})
```

```python
dfl = pd.melt(population[["Année", "nb_loups_tués", "Esti_OFB"]], ["Année"])
popl = sns.lineplot(data=dfl, x="Année", y="value", hue="variable");
popl.set_xticks(range(1995, 2025));
popl.set_yticks(range(0, 1500, 250));
popl.set_xticklabels(popl.get_xticklabels(), rotation='vertical');
```

```python
popl = sns.barplot(data=population.loc[(population["Année"] > 2014)], x="Année", y="pct_tués");
# popl.set_xticks(range(1995, 2025));
popl.set_xticklabels(popl.get_xticklabels(), rotation='vertical');
```

```python
dfl = pd.melt(population[["Année", "pct_tués", "pct_diff"]].loc[(population["Année"] > 2014)], ["Année"])
popl = sns.lineplot(data=dfl, x="Année", y="value", hue="variable");
popl.set_xticks(range(2015, 2025));
popl.set_xticklabels(popl.get_xticklabels(), rotation='vertical');
```
